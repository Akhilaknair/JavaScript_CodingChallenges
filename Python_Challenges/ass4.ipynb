```python
#q1
import numpy as np

arr = np.array([[16, 3, 5],
                [1, 7, 2]])

print("Original Array:\n", arr)

sorted_axis0 = np.sort(arr, axis=0)
print("\nSorted along the first axis (axis=0):\n", sorted_axis0)

sorted_axis_last = np.sort(arr, axis=-1)
print("\nSorted along the last axis (axis=-1):\n", sorted_axis_last)

sorted_flattened = np.sort(arr, axis=None)
print("\nSorted flattened array:\n", sorted_flattened)
```

```python
import numpy as np

arr = np.array(['python', 'CODING', 'nUmPy', 'pAnDas'])

print("Original Array:")
print(arr)

capitalized = np.char.capitalize(arr)
print("\nCapitalized:")
print(capitalized)

lowercase = np.char.lower(arr)
print("\nLowercase:")
print(lowercase)

uppercase = np.char.upper(arr)
print("\nUppercase:")
print(uppercase)

swapcase = np.char.swapcase(arr)
print("\nSwapcase:")
print(swapcase)


titlecase = np.char.title(arr)
print("\nTitle-case:")
print(titlecase)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'id': [4, 5],
    'name': ['David', 'Eva']
})

joined_df = pd.concat([df1, df2], ignore_index=True)
print("Joined DataFrame along rows:")
print(joined_df)

df3 = pd.DataFrame({
    'id': [1, 2, 4, 5],
    'score': [85, 90, 88, 92]
})

merged_df = pd.merge(joined_df, df3, on='id')
print("\nMerged DataFrame on 'id':")
print(merged_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample sales data
data = {
    'Date': pd.date_range(start='2023-01-01', periods=6, freq='M'),
    'Product': ['A', 'B', 'A', 'C', 'B', 'C'],
    'Units_Sold': [120, 150, 180, 100, 200, 140],
    'Revenue': [2400, 3000, 3600, 2000, 4000, 2800]
}

df = pd.DataFrame(data)

print("Sales Data:")
print(df)


#line plot 
plt.figure(figsize=(8, 4))
plt.plot(df['Date'], df['Revenue'], marker='o', color='green')
plt.title('Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.grid(True)
plt.tight_layout()
plt.show()

#bar plot
plt.figure(figsize=(6, 4))
sns.barplot(x='Product', y='Units_Sold', data=df, estimator=sum)
plt.title('Total Units Sold per Product')
plt.tight_layout()
plt.show()

#pie chart 
revenue_by_product = df.groupby('Product')['Revenue'].sum()

plt.figure(figsize=(5, 5))
plt.pie(revenue_by_product, labels=revenue_by_product.index, autopct='%1.1f%%', startangle=140)
plt.title('Revenue Share by Product')
plt.show()
```

```python
import tkinter as tk
from tkinter import messagebox
from datetime import datetime

def calculate_age():
    try:
        birth_date = entry.get()
        birth_date = datetime.strptime(birth_date, "%Y-%m-%d")
        today = datetime.today()
        age = today.year - birth_date.year - (
            (today.month, today.day) < (birth_date.month, birth_date.day)
        )
        messagebox.showinfo("Age", f"Your age is: {age} years")
    except ValueError:
        messagebox.showerror("Invalid input", "Please enter correct format.")

root = tk.Tk()
root.title("Age Calculator")
root.geometry("300x200")

tk.Label(root, text="Age Calculator", font=("Arial", 16)).pack(pady=10)

tk.Label(root, text="Enter your Date of Birth (YYYY-MM-DD):").pack()

entry = tk.Entry(root, width=25)
entry.pack(pady=5)

tk.Button(root, text="Calculate Your Age", command=calculate_age).pack(pady=10)

root.mainloop()
```
